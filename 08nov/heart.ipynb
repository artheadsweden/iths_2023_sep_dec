Split the dataset into independent and dependent variables

Independent variables are all the columns except hd

Dependent variable is the hd column

```python
# Extract independent variables
X = df.drop('hd', axis=1).copy()
X.head()
```

```python
# Extract dependent variable
y = df['hd'].copy()
y.head()
```

We can see that the columns sex, cp, fbs, restecg, exang, slope, ca and thal are categorical. We need to encode them.

```python
# Check the categorical variables
check_columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
for col in check_columns:
    print(col, X[col].unique())
```

One hot encoding
We can do this with ColumnTransformer from sklearn or get_dummies from pandas.

```python
# See how it works
pd.get_dummies(X, columns=['cp']).head()
```

```python
X_encoded = pd.get_dummies(X, columns=['cp', 'restecg', 'slope', 'ca', 'thal'])
```

```python
X_encoded.head()
```

```python
X_encoded.dtypes
```

Lets make our y binary

```python
y.unique()
```

```python
# Get the index of the rows with the value >= 1
y_not_zero_index = y > 0
```

```python
y_not_zero_index
```

```python
# Set the value of the rows with the index to 1
y[y_not_zero_index] = 1
```

```python
y.unique()
```

Make our initial decision tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=42)
clf_dt = DecisionTreeClassifier(random_state=42)
clf_dt.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 7.5))

plot_tree(
    clf_dt, 
    filled=True, 
    rounded=True, 
    class_names=['No Disease', 'Disease'], 
    feature_names=X_encoded.columns
)
```

Plot the result as a confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
predictions = clf_dt.predict(X_test)
cm = confusion_matrix(y_test, predictions, labels=[0, 1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Disease', 'Disease'])
disp.plot()
plt.show()
```

Prune the tree using cost complexity pruning

```python
paths = clf_dt.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = paths.ccp_alphas
ccp_alphas = ccp_alphas[:-1]
ccp_alphas
```

```python
clf_dts = []
for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    clf_dt.fit(X_train, y_train)
    clf_dts.append(clf_dt)
```

```python
traning_scores = [clf_dt.score(X_train, y_train) for clf_dt in clf_dts]
test_scores = [clf_dt.score(X_test, y_test) for clf_dt in clf_dts]
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('alpha')
ax.set_ylabel('accuracy')
ax.set_title('Accuracy vs alpha for training and testing sets')
ax.plot(ccp_alphas, traning_scores, marker='o', label='train', drawstyle='steps-post')
ax.plot(ccp_alphas, test_scores, marker='o', label='test', drawstyle='steps-post')
ax.legend()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

clf_dt = DecisionTreeClassifier(random_state=42, ccp_alpha=0.016)
scores = cross_val_score(clf_dt, X_train, y_train, cv=5)

df = pd.DataFrame(data={'tree': range(5), 'accuracy': scores})
```

```python
df
```

```python
df.plot(x='tree', y='accuracy', marker='o', linestyle='--')
```

```python
import numpy as np

aplha_loop_scores = []
for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
    aplha_loop_scores.append([ccp_alpha, np.mean(scores), np.std(scores)])
```

```python
alpha_results = pd.DataFrame(aplha_loop_scores, columns=['alpha', 'mean_accuracy', 'std'])
alpha_results.plot(x='alpha', y='mean_accuracy', yerr='std', marker='o', linestyle='--')
```

```python
alpha_results[(alpha_results['alpha'] > 0.012) & (alpha_results['alpha'] < 0.015)]
```

```python
optimal_ccp_alpha = alpha_results[(alpha_results['alpha'] > 0.012) & (alpha_results['alpha'] < 0.015)]['alpha']
```

```python
optimal_ccp_alpha = float(optimal_ccp_alpha)
```

```python
optimal_ccp_alpha
```

Make the final decision tree with the best alpha

```python
clf_dt_pruned = DecisionTreeClassifier(random_state=42, ccp_alpha=optimal_ccp_alpha)
```

```python
clf_dt_pruned.fit(X_train, y_train)
```

```python
plt.figure(figsize=(15, 7.5))

plot_tree(
    clf_dt_pruned, 
    filled=True, 
    rounded=True, 
    class_names=['No Disease', 'Disease'], 
    feature_names=X_encoded.columns
)
```

```python
predictions = clf_dt_pruned.predict(X_test)
cm = confusion_matrix(y_test, predictions, labels=[0, 1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Disease', 'Disease'])
disp.plot()
plt.show()
```
