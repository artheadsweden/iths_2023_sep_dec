# Elbow method
Run K-means Clustering: Perform K-means clustering on your dataset for a range of cluster numbers (e.g., k = 1, 2, 3, ..., 10). For each k, the K-means algorithm will partition the dataset into k clusters.

Calculate Within-Cluster Sum of Squares (WCSS): For each k, calculate the within-cluster sum of squares (WCSS). WCSS is a measure of variance within each cluster, and it's calculated as the sum of the squared distances between each data point and the centroid of its assigned cluster. The goal of K-means is to minimize this value.

Plot the Elbow Curve: Create a plot with the number of clusters on the x-axis and the corresponding WCSS on the y-axis. This plot is often referred to as the "elbow curve."

Identify the Elbow Point: As you increase the number of clusters, the WCSS tends to decrease because the data points are closer to the centroids they are assigned to. The rate of decrease sharply changes at a certain point, creating an "elbow" in the graph. This point is typically considered to be an indicator of the appropriate number of clusters.

Interpretation: The idea is that before the elbow, adding more clusters significantly decreases the WCSS, meaning that the clustering is gaining more definition. Beyond the elbow point, improvements in WCSS are less substantial. Hence, the number of clusters at the elbow is considered a good balance between the number of clusters and the variance explained.

```python
from sklearn.cluster import KMeans

wcss = [KMeans(n_clusters=i).fit(X).inertia_ for i in range(1, 11)]
wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
pred_y = kmeans.fit_predict(X)  
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
plt.show()
```
